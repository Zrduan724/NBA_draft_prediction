```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import plot_confusion_matrix, accuracy_score, precision_score, mean_squared_error
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.ensemble import AdaBoostClassifier
```

```python
raw_data_path = r".\dataset\collegeplayer_cleaned.csv"
raw_data = pd.read_csv(raw_data_path)
raw_data.head()
```

```python
index_list = []

for i in range(len(raw_data)-1):
    if raw_data['player_name'][i] == raw_data['player_name'][i+1]:
        index_list.append(i)

new_data = raw_data.drop(index=index_list, axis=0)
new_data = new_data.reset_index(drop=True) # 重新排序
new_data
```

```python
delet_index = [0, 2, 3, 6, 8, 9, 10, 32, 33, 34, 35, 36,
               37, 38, 39, 40, 71, 72]
drop_list = []

for index in delet_index:
    drop_list.append(raw_data.columns[index])
    
print(drop_list)
```

```python
new_data_cleaned = new_data.drop(columns=drop_list)
new_data_cleaned
```

## impute missing data

```python
new_data_cleaned.drop(columns=['player_name', 'YEAR', 'OVERALL'], inplace=True)
new_data_cleaned
```

```python
# Filling missing data
si = SimpleImputer(strategy='median')
new_data_cleaned_imputed = si.fit_transform(new_data_cleaned)
print(new_data_cleaned_imputed.shape)
np.save('./NCAA_data_cleaned_imputed_twoclass.npy', new_data_cleaned_imputed)
```

```python
array_2021 = new_data_cleaned_imputed[:48, :]
array_rest = new_data_cleaned_imputed[48:, :]
# print(array_2021.shape)
# print(array_rest.shape)
np.save('./data_2021_twoclass.npy', array_2021)
np.save('./data_rest_twoclass.npy', array_rest)
```
