```python
import numpy as np
import pandas as pd
```

```python
data_path = r'./NCAA_data_cleaned.csv'
data = pd.read_csv(data_path)
data.head()
```

```python
data.drop(columns=['GP'], inplace=True) # inplace=true会自动重排序
data.head()
```

```python
# data.columns[1: 3]
scale_list = []
for index in range(3, 48):
    scale_list.append(data.columns[index])
print(scale_list)
```

```python
# 对需要归一化的列进行归一化
data[scale_list] = data[scale_list].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
data.head()
```

```python
data.to_csv('./normalized_data.csv', index=False)
```
