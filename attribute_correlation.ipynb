## Split dataset into test_data and train_data

```python
data_path = r"./NCAA_data_cleaned_v2_inputed.npy"
data_array = np.load(data_path)
print(data_array.shape)
data_array
```

```python
array_2021 = data_array[:48, :]
array_rest = data_array[48:, :]
# print(array_2021.shape)
# print(array_rest.shape)
np.save('./data_2021.npy', array_2021)
np.save('./data_rest.npy', array_rest)
```

```python
test_array = np.zeros(shape=[84, 46])
test_index_list = []

for i in range(data_array.shape[0]-1):
    if data_array[i][-1] != data_array[i+1][-1]:
        test_index_list.append(i)

test_index_list.append(data_array.shape[0]-1)
print(len(test_index_list))
print(test_index_list)
```

```python
import random
```

```python
random_items = random.sample(test_index_list, 84)
print(random_items)
```

```python
count = 0
for row in random_items:
    test_array[count] = data_array[row]
    count = count + 1

train_array = np.delete(data_array, random_items, axis=0)
print(test_array)
print(train_array)
print(data_array.shape)
print(test_array.shape)
print(train_array.shape)
```

```python
np.save('./test_data.npy', test_array)
np.save('./train_data.npy', train_array)
```
